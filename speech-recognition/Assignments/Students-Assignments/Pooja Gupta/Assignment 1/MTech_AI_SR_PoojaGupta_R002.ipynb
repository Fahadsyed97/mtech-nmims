```python
from gtts import gTTS
import os
import IPython.display as ipd
```

```python
text1 = 'Sesquipedalian'
text2 = 'Perspicacity'
text3 = 'Ostentation'
text4 = 'Hippopotomonstrosesquipedaliophobia'
text5 = 'Floccinaucinihilipilification'
text6 = 'Scripturient'
language='en'
```

```python
def HumanSpeechProduction(text):
  tts_object = gTTS(text=text,lang=language,slow=False)
  tts_object.save('{}.wav'.format(text))
  ipd.Audio('{}.mp3'.format(text))
```

```python
ipd.Audio('{}.mp3'.format(text5))
```

```python
text_list = [text1,text2,text3,text4,text5,text6]
for i,text in enumerate(text_list):
  print(text)
  HumanSpeechProduction(text)
  
```

## Analyzing the text file


```python
!pip install librosa
```

```python

import matplotlib.pyplot as plt
import librosa.display
```

```python
import librosa
audio_path1 = "Sesquipedalian.wav"
audio_path2 = "Perspicacity.wav"
audio_path3 = "Ostentation.wav"
audio_path4 = "Hippopotomonstrosesquipedaliophobia.wav"
audio_path5 = "Floccinaucinihilipilification.wav"
audio_path6 = "Scripturient.wav"
```

```python
audio = [audio_path1,audio_path2,audio_path3,audio_path4,audio_path5,audio_path6]
for i,audio in enumerate(audio):
  x , sr = librosa.load(audio)


  plt.figure(figsize=(14, 5))
  librosa.display.waveplot(x, sr=sr)

  X = librosa.stft(x)
  Xdb = librosa.amplitude_to_db(abs(X))

  plt.figure(figsize=(14, 5))
  librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='hz')
  plt.colorbar()
  plt.xlabel(audio)
```

#### Inference:
* The redness in the spectogram above corresponding to lower Hz represents data having low pitch and corresponding to higher Hz represents data having higher pitch.
