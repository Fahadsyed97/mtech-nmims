```python
from gtts import gTTS
import IPython.display as ipd
import librosa
import librosa.display
import os 
import warnings
import matplotlib.pyplot as plt

warnings.filterwarnings("ignore")
```

```python
texts = ["Sesquipedalian","Perspicacity","Ostentation","Muliebrity","Hippopotomonstrosesquipedaliophobia","Floccinaucinihilipilification","Scripturient"]
language='en'
```

```python
for text in texts:
    gtts_object = gTTS(text=text,lang=language,slow=False)
    gtts_object.save(f"{text}.wav")
```

```python
def show_waveplot(audio_path):
  x , sr = librosa.load(audio_path)
  plt.figure(figsize=(20, 6))
  librosa.display.waveshow(x, sr=sr)

def show_spectrogram(audio_path):
  x , sr = librosa.load(audio_path)
  X = librosa.stft(x)
  Xdb = librosa.amplitude_to_db(abs(X))
  plt.figure(figsize=(20,6))
  librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='hz')
  plt.colorbar()
```

```python
ipd.Audio("Sesquipedalian.wav")
```

### Waveplot
- The amplitude differences can be seen in the waveplot for different parts of the word which are emphasized on while pronouncing it
- Ses, qui, ped, dalian - the loudness for this separate terms are indicated in the amplitudes of the waveplot

```python
show_waveplot("Sesquipedalian.wav")
```

### Spectrogram
- Both frequency and time domains can be analyzed using the spectrogram
- The colors represent the magnitude of the observed frequencies 
- The smaller frequencies (0-3kHZ) have much stronger magnitude
- The frequency during the inital part of the word 'Sesque' is relatively lower (0-1kHz) than the middle part of the word (0-3kHz)

```python
show_spectrogram("Sesquipedalian.wav")
```
